```python
import pandas as pd
import numpy as np

# Load the first dataset
data_1 = pd.read_csv("/Users/leo/Desktop/Projects/Data Log Analysis/Data/CCJ 1mm sim Bando red Wanbao 1100 8g.csv", skiprows=15)

# Store original column names before cleaning
original_columns_1 = set(data_1.columns)

# Remove the first row and columns with all NaN values
data_1 = data_1.drop(index=0)
data_1 = data_1.dropna(axis=1, how="all").loc[:, (data_1 != 0.0).any()]
data_1 = data_1.drop(columns=['Vehicle_Speed'])
data_1 = data_1.drop(columns=['Dec_FC_En'])


# Function to remove constant columns
def remove_constant_columns(df):
    return df.loc[:, df.nunique() > 1]

# Apply the function
data_1 = remove_constant_columns(data_1)

# Identify dropped columns
dropped_columns_1 = original_columns_1 - set(data_1.columns)

# Print the dropped columns
print("Dropped columns from data_1:", dropped_columns_1)


# Load the second dataset
data_2 = pd.read_csv("/Users/leo/Desktop/Projects/Data Log Analysis/Data/CCJ 1mm sim Bando red Wanbao 1100 8+9g.csv", skiprows=15)

# Store original column names before cleaning
original_columns_2 = set(data_2.columns)

# Remove the first row and columns with all NaN values
data_2 = data_2.drop(index=0)
data_2 = data_2.dropna(axis=1, how="all").loc[:, (data_2 != 0.0).any()]
data_2 = data_2.drop(columns=['Vehicle_Speed'])
data_2 = data_2.drop(columns=['Dec_FC_En'])

# Apply the function
data_2 = remove_constant_columns(data_2)

# Identify dropped columns
dropped_columns_2 = original_columns_2 - set(data_2.columns)

# Print the dropped columns
print("Dropped columns from data_2:", dropped_columns_2)
```

```python
col_name = 'TPS_Percent'  # Replace with your actual column name

# Find the first index where the column value is non-zero
first_nonzero_index = (data_1[col_name] > 1).idxmax()

# Slice the data from the first non-zero occurrence
data_after_first_nonzero = data_1.loc[first_nonzero_index:]

# Find the first occurrence of 0.0 after the initial non-zero value
first_zero_after_nonzero_index = data_after_first_nonzero[data_after_first_nonzero[col_name] == 0.0].index.min()

# Extract the section of data between first nonzero and when it returns to 0.0
if pd.notna(first_zero_after_nonzero_index):  # Ensure a valid index is found
    data_section_1 = data_1.loc[first_nonzero_index-10:first_zero_after_nonzero_index+10].reset_index(drop=True)
    print("Extracted data section:")
    print(data_section_1)
else:
    print("No return to 0.0 found. Using data from first nonzero onward.")
    data_section_1 = data_1.loc[first_nonzero_index:].reset_index(drop=True)

# Save the extracted section to a CSV
data_section_1['TPS_Percent']
```

```python
import plotly.graph_objects as go

# Assuming 'data_section_2' is your DataFrame
x_col = 'Time'   # Column for X-axis
y_col1 = 'RPM'   # First column for Y-axis (range from 2000 to 10000)
y_col2 = 'PH_GPS_Speed'  # Second column for Y-axis (range from 0 to 100)
y_col3 = 'TPS_Percent'

# Create a figure
fig = go.Figure()

# Add the first plot (RPM) with the first Y-axis
fig.add_trace(go.Scatter(
    x=data_section_1[x_col], 
    y=data_section_1[y_col1], 
    mode='lines',  # 'lines+markers' for both lines and markers
    name=y_col1,  # Name for the legend
    line=dict(color='green', width=2),  # Line color and width
    marker=dict(symbol='circle', size=6)  # Marker style
))

# Add the second plot (PH_GPS_Speed) with the second Y-axis
fig.add_trace(go.Scatter(
    x=data_section_1[x_col], 
    y=data_section_1[y_col2], 
    mode='lines',  # 'lines+markers' for both lines and markers
    name=y_col2,  # Name for the legend
    line=dict(color='blue', width=2),  # Line color and width
    marker=dict(symbol='x', size=6),  # Marker style
    yaxis='y2'  # This specifies that the second axis should be used
))

# Add the third plot (Another_Column) with the third Y-axis
fig.add_trace(go.Scatter(
    x=data_section_1[x_col], 
    y=data_section_1[y_col3], 
    mode='lines',  # 'lines+markers' for both lines and markers
    name=y_col3,  # Name for the legend
    line=dict(color='red', width=2),  # Line color and width
    marker=dict(symbol='square', size=6),  # Marker style
    yaxis='y3'  # This specifies that the third axis should be used
))

# Update layout for better presentation
fig.update_layout(
    title="Comparison of RPM, GPS Speed, and Another Column",  # Title of the plot
    xaxis_title=x_col,  # X-axis label
    yaxis_title=y_col1,  # Left Y-axis label for RPM
    yaxis2=dict(
        title=y_col2,  # Right Y-axis label for PH_GPS_Speed
        overlaying='y',  # Share the same X-axis
        side='right'  # Place on the right side
    ),
    yaxis3=dict(
        title=y_col3,  # Right Y-axis label for Another_Column
        overlaying='y',  # Share the same X-axis
        side='right',  # Place on the right side (can adjust to left if needed)
        position=0.85  # Adjust position if needed to avoid overlap
    ),
    legend=dict(x=0.01, y=0.95, font=dict(size=8)),  # Position legend
    template='plotly_dark',  # Optional: Dark theme for the graph
    width=1200,  # Set width of the plot
    height=1000  # Set height of the plot
)

# Show the interactive plot
fig.show()
```
